```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
imagem=cv2.imread("madfox.jpg")
cap = cv2.VideoCapture(0)
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)

surfdec = cv2.xfeatures2d.SURF_create(hessianThreshold=5000)
kp1, des1 = surfdec.detectAndCompute(imagem,None)

def detect_features(img,kp1,des1,frame,frame_g):
    MIN_MATCH_COUNT = 10
   
    
    kp2, des2 = surfdec.detectAndCompute(frame_g,None)

    FLANN_INDEX_KDTREE = 0
    index_params = dict(algorithm = FLANN_INDEX_KDTREE, trees = 5)
    search_params = dict(checks = 50)

    flann = cv2.FlannBasedMatcher(index_params, search_params)

    matches = flann.knnMatch(des1,des2,k=2)

    good = []
    for x,y in matches:
        if x.distance < 0.7*y.distance:
            good.append(x)

    circles = None
    circles=cv2.HoughCircles(frame_g,cv2.HOUGH_GRADIENT,2,60,param1=200,param2=100,minRadius=5,maxRadius=60)

    if circles is not None:
        circles = np.uint16(np.around(circles))

        for i in circles[0,:]:
            print(i)
            cv2.circle(frame,(i[0],i[1]),i[2],(0,255,0),2)
            cv2.circle(frame,(i[0],i[1]),2,(0,0,255),3)

    if len(good)>MIN_MATCH_COUNT and circles is not None:
        src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
        dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)


        M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)
        matchesMask = mask.ravel().tolist()
        t,w = img.shape[0],img.shape[1]
        pts = np.float32([ [0,0],[0,t-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)

        dst = cv2.perspectiveTransform(pts,M)

        frame = cv2.polylines(frame,[np.int32(dst)],True,255,3, cv2.LINE_AA)
        
    else:
        print("Not enough matches are found - %d/%d" % (len(good),MIN_MATCH_COUNT))
        matchesMask = None

    draw_params = dict(matchColor = (0,255,0), # draw matches in green color
                    singlePointColor = None,
                    matchesMask = matchesMask, # draw only inliers
                    flags = 2)
  

    return frame


while(True):

    ret, frame = cap.read()

    frame_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    frame_gray=cv2.medianBlur(frame_gray,5)
  

    frame=detect_features(imagem,kp1, des1,frame,frame_gray)
    
    cv2.imshow('original',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
